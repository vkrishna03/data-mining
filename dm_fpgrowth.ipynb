```python
import numpy as np
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules, apriori ,fpgrowth
```

```python
data = [
    ['I1','I2','I5'],
    ['I2','I4'],
    ['I2','I3'],
    ['I1','I2','I4'],
    ['I1','I3'],
    ['I2','I3'],
    ['I1','I3'],
    ['I1','I2','I3','I5'],
    ['I1','I2','I3']
]
```

```python
te = TransactionEncoder()
te_ary = te.fit(data).transform(data)
df=pd.DataFrame(te_ary,columns=te.columns_)
df
```

```python
fpgrowth(df,min_support=0.2,use_colnames=True)
```

```python
%timeit -n 100 -r 10 apriori(df,min_support=0.2,use_colnames=True)
```

```python
%timeit -n 100 -r 10 fpgrowth(df,min_support=0.2,use_colnames=True)
```
