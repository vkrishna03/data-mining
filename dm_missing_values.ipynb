```python
import pandas as pd

path = "dm-missing-values-sheets.csv"

df = pd.read_csv(path)
df
```

```python
df["quantity"] = df["quantity"].fillna(df["quantity"].mean())
df
```

```python
df["price"] = df["price"].fillna(df["price"].median())
df
```

```python
df["bought"] = df["bought"].fillna(df["bought"].std())
df
```

```python
df["forenoon"] = df["forenoon"].fillna(df["forenoon"].min())
df
```

```python
df["afternoon"] = df["afternoon"].fillna(df["afternoon"].max())
df
```
